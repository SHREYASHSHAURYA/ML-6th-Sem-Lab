```python
import numpy as np

# Features
X = np.array([
    [1, 50],
    [2, 60],
    [3, 65],
    [4, 70],
    [5, 80]
])

# Target
y = np.array([3, 4, 5, 6, 8])
```

```python
# Add column of 1s for intercept
ones = np.ones((X.shape[0], 1))
X_b = np.hstack((ones, X))
```

```python
# Normal Equation
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Intercept and Coefficients:")
print(beta)
```

```python
# New input (with bias term)
new_person = np.array([1, 3, 75])

salary_pred_ml = new_person @ beta

print("Multilinear Prediction:", salary_pred_ml)
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))
```

```python
def knn_predict(X, y, query, k=3):
    distances = []

    for i in range(len(X)):
        dist = euclidean_distance(X[i], query)
        distances.append((dist, y[i]))

    # Sort by distance
    distances.sort(key=lambda x: x[0])

    # Take k nearest
    neighbors = distances[:k]

    # Average their salaries
    avg_salary = np.mean([n[1] for n in neighbors])

    return avg_salary
```

```python
query = np.array([3, 75])

salary_pred_knn = knn_predict(X, y, query, k=3)

print("KNN Prediction:", salary_pred_knn)
```
