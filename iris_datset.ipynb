```python
import pandas as pd
from sklearn.datasets import load_iris

# 1. Load the Iris dataset
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target_names[iris.target]

print("1. Iris Dataset Loaded:")
display(df_iris.head())
```

```python
# 2. Display basic information about the DataFrame
print("\n2. DataFrame Info:")
df_iris.info()
```

```python
# 3. Display descriptive statistics
print("\n3. Descriptive Statistics:")
display(df_iris.describe())
```

```python
# 4. Check unique values in the 'species' column
print("\n4. Unique Species:")
display(df_iris['species'].unique())
```

```python
# 5. Group by 'species' and calculate the mean of numerical features
print("\n5. Mean of features grouped by Species:")
display(df_iris.groupby('species').mean())
```

```python
# 6. Filter data for a specific species (e.g., 'setosa')
print("\n6. Filtered data for 'setosa' species:")
df_setosa = df_iris[df_iris['species'] == 'setosa']
display(df_setosa.head())
```

```python
# 7. Add a new calculated column: 'sepal_area'
df_iris['sepal_area'] = df_iris['sepal length (cm)'] * df_iris['sepal width (cm)']
print("\n7. DataFrame with 'sepal_area' column:")
display(df_iris.head())
```

```python
# 8. Sort the DataFrame by 'sepal_area' in descending order
print("\n8. DataFrame sorted by 'sepal_area' (top 5 rows):")
df_sorted = df_iris.sort_values(by='sepal_area', ascending=False)
display(df_sorted.head())
```

```python
# 9. Export the modified DataFrame to a new CSV file
output_filename_iris = 'iris_modified_data.csv'
df_iris.to_csv(output_filename_iris, index=False)
print(f"\n9. Modified Iris DataFrame exported to '{output_filename_iris}'")
```

```python
# 10. Import the exported CSV back into a new DataFrame
df_reimported_iris = pd.read_csv(output_filename_iris)
print(f"\n10. Modified Iris DataFrame re-imported from '{output_filename_iris}':")
display(df_reimported_iris.head())
```
