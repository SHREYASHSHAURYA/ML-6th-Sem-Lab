```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.xlabel("Sepal")
plt.ylabel("Petal")
plt.show()
```
